```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df_produtos = pd.read_csv('datasets_olist/olist_products_dataset.csv')
df_produtos
```

**Tabela Produto**
* product_id: código do produto
* product_category_name: categoria
* product_name_lenght: Nome e comprimento do produto
* product_description_lenght: descrição e comprimento do produto
* product_photos_qty: quantidade de fotos do produto
* product_weight_g: peso do produto
* product_length_cm: comprimento
* product_height_cm: altura
* product_width_cm: largura

```python
df_itens_pedido = pd.read_csv('datasets_olist/olist_order_items_dataset.csv')
df_itens_pedido
```

**Tabela de itens dos pedidos**

* order_id: Identificador do pedido
* product_id: Identificador do produto
* seller_id: Identificador do vendedor
* shipping_limit_date: timestamp limite para a entrega do pedido
* price: valor do pedido
* freight_value: valor do frete

```python
df_pagamentos = pd.read_csv('datasets_olist/olist_order_payments_dataset.csv')
df_pagamentos
```

**Tabela de tipo de pagamentos**

* order_id: identificador do pedido
* payment_sequential: (não ficou claro essa coluna)
* payment_type: tipo de pagamento
* payment_installments: parcelas do pagamento
* payment_value: valor que foi pago


## Perguntas

* Qual categoria de produto é mais cara?
* Qual categoria de produto é mais barata?
* Qual é a forma de pagamento menos utilizada pelos consumidores?
* Qual categoria os consumidores mais compraram?



```python
df_olist = pd.merge(df_produtos, df_itens_pedido, on='product_id')
df_olist
```

```python
df_olist = pd.merge(df_olist, df_pagamentos, on='order_id')
df_olist
```

```python
df_olist.to_csv('base_final_analise_olist.csv', index=False)
```

# Analise de dados

## Respondendo as perguntas

* 1.Qual categoria de produto é mais cara?
* 2.Qual categoria de produto é mais barata?
* 3.Qual é a forma de pagamento menos utilizada pelos consumidores?
* 4.Qual a quantidade de maxima de parcelas a ser feita?


```python
df_itens_caros = df_olist.sort_values(by='price', ascending=False).head(10)
df_itens_caros
```

```python
top10_produtos_caros = df_itens_caros.sort_values(by='price', ascending=False).head(10)


plt.figure(figsize=(15, 10))
plt.bar(top10_produtos_caros['product_category_name'],top10_produtos_caros['price'], color = 'pink')
plt.xlabel('Categorias de Produtos')
plt.ylabel('Preço')
plt.title('Categorias com os maiores preços')
```

1.Utilidades domesticas é a categoria mais cara.

```python
df_itens_baratos = df_olist.sort_values(by='price', ascending=True).head(10)
df_itens_baratos
```

```python
top10_produtos_baratos = df_itens_baratos.sort_values(by='price', ascending=True).head(10)


plt.figure(figsize=(10, 6))
plt.bar(top10_produtos_baratos['product_category_name'],top10_produtos_baratos['price'], color = 'grey')
plt.xlabel('Categorias de Produtos')
plt.ylabel('Preço')
plt.title('Categorias com menores preços')
```

2.Ferramentas de construção é a categoria mais barata.

```python
df_olist['payment_type'].value_counts()
```

3.'Crédito' é o mais usado entre os consumidores.
